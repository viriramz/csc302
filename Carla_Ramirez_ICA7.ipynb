```python
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd
import matplotlib.pyplot as plt

file_path = "/content/drive/My Drive/DATA/boxoffice.csv"

boxoffice = pd.read_csv(file_path)

print("First 5 rows of the dataset:")
print(boxoffice.head())

star_wars_movies = boxoffice[boxoffice['title'].str.contains("Star Wars", case=False, na=False)]

print("\nBox office earnings for Star Wars movies:")
print(star_wars_movies[['title', 'amount']])

boxoffice['amount'].plot.bar()
plt.title("Box Office Earnings (Pandas Plot)")
plt.xlabel("Movie Index")
plt.ylabel("Amount")
plt.show()

plt.figure(figsize=(15, 3))
plt.bar(boxoffice['title'], boxoffice['amount'])
plt.xticks(rotation=45, ha="right")  # Rotate x labels for readability
plt.title("Box Office Earnings (Matplotlib Bar Chart)")
plt.xlabel("Movies")
plt.ylabel("Amount")
plt.show()
```
